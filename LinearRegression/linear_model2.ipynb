```python
def getnull_count(colname):
    count=0
    for val in data[colname]:
        if val==0:
            count=count+1
    return count
```

```python
def get_median_without_null(colname):
    not_null_yr_renovated = []
    for val in data[colname]:
        if val!=0:
            not_null_yr_renovated.append(val)
    fram = pd.DataFrame(not_null_yr_renovated, columns=['values'])
    med_yr = fram.median()
    return med_yr
```
